```python
from fastai.vision.all import *
from fastai.vision.widgets import *
import urllib.request

from fastai.imports import *
from fastai import *
#from fastai.vision import *
#from fastai.core import *
from fastai.vision.data import ImageDataLoaders
```

```python
path = Path(r"C:\Users\nandini.singh\Desktop\fastai\data")
data = ImageDataLoaders.from_folder(path, valid_pct=0.3,item_tfms = RandomResizedCrop(128, min_scale=0.35))
#data = ImageDataBunch.from_folder(path,test='test', ds_tfms=get_transforms(), size=360)
data.show_batch(nrows=5, figsize=(10,20))
```

```python
learn = cnn_learner(data, resnet18, metrics=partial(accuracy))
```

```python
learn.fine_tune(5, base_lr=3e-3)
```

```python
preds,targs = learn.get_preds()
```

```python
learn.export("new_model")
```

```python
learn.export("export.pkl")
```

```python
import matplotlib.pyplot as plt
from IPython.display import display
from PIL import Image
from keras.preprocessing.image import load_img
```

```python
btn_upload = widgets.FileUpload()
out_pl = widgets.Output()
lbl_pred = widgets.Label()
```

```python
learn_inf = load_learner('export.pkl', cpu=True)
```

```python
# Create interface

def on_data_change(change):
    lbl_pred.value = ''
    
    # Load Image
    img = PILImage.create(btn_upload.data[-1])
    out_pl.clear_output()
    with out_pl:
        display(img.to_thumb(224,244))
    
    # Predict image
    pred, pred_idx, probs = learn_inf.predict(img)
    
    lbl_pred.value = f'Prediction: {pred}; Probability: {probs[pred_idx]:.04f}'
```

```python
btn_upload.observe(on_data_change, names=['data'])

# Display Predictions
display(VBox([widgets.Label('Select your Image!'), btn_upload, out_pl, lbl_pred]))
```
